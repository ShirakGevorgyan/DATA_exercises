```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in df.columns:
    if df[i].dtype == object:
        lable = LabelEncoder()
        df[i] = lable.fit_transform(df[i])
        
```

```python
df
```

```python
df['veil-type'].value_counts()
```

```python
mec_list = []
for i in df.columns:
    mijin = df[i].mean()
    poqr_list = []
    hamarich = 0
    for j in df[i]:
        hamarich += (j-mijin)**2
    for j in df[i]:
        poqr_list.append((j - mijin)/(hamarich/len(df[i]))**0.5)
    mec_list.append(poqr_list)
        
```

```python
mec_list
```

```python
df.corr()['class']
```

```python
X = df.drop(['cap-shape', 'cap-color', 'veil-type'] , axis=1)
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=66)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestClassifier()
```

```python
params = {
    'n_estimators': [75, 100, 150],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 15],
    'min_samples_split': [2, 3, 5],
    'min_samples_leaf': [1, 2, 3],
    'min_weight_fraction_leaf': [0.0, 0.1]
}
```

```python
full_model = GridSearchCV(model, params, cv=3, verbose=2)
```

```python
full_model.fit(X_train, y_train)
```

```python
full_model.best_params_
```

```python
y_pred = full_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay , accuracy_score,  precision_score
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_pred, y_test)).plot()
```
